```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
```

```python
# 1. Baca gambar (sebagai contoh, pastikan kamu punya file gambar grayscale)
# Matriks akan diubah ke tipe data integer agar bisa ditambah/kurang
gambar_asli = mpimg.imread('gambar_citra.jfif')

# Jika gambar RGB, konversi dulu ke array grayscale secara manual atau ambil 1 channel saja
if len(gambar_asli.shape) == 3:
    gambar_asli = np.mean(gambar_asli, axis=2).astype(int)
else:
    gambar_asli = gambar_asli.astype(int)

# Konversi numpy array ke list Python biasa agar murni tanpa library komputasi
citra_A = gambar_asli.tolist()
```

```python
# Fungsi murni untuk pencerahan gambar grayscale (1 channel)
def manual_image_brightening(citra_A, b):
    # Mendapatkan ukuran resolusi citra M x N
    M = len(citra_A)      # baris (tinggi)
    N = len(citra_A[0])   # kolom (lebar)

    # Membuat matriks kosong (citra B) dengan ukuran yang sama
    citra_B = [[0 for _ in range(N)] for _ in range(M)]

    # Melakukan perulangan untuk setiap piksel (mirip algoritma di PDF)
    for i in range(M):
        for j in range(N):
            # Menambahkan skalar b ke nilai piksel
            temp = citra_A[i][j] + b

            # Operasi Clipping
            if temp < 0:
                citra_B[i][j] = 0
            elif temp > 255:
                citra_B[i][j] = 255
            else:
                citra_B[i][j] = temp

    return citra_B
```

```python
# 2. Proses pencerahan (misal ditambah 100 seperti di contoh PDF)
nilai_b = 100
citra_B = manual_image_brightening(citra_A, nilai_b)
```

```python
# 3. Tampilkan Hasil
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(citra_A, cmap='gray', vmin=0, vmax=255)
axes[0].set_title('Citra Semula')

axes[1].imshow(citra_B, cmap='gray', vmin=0, vmax=255)
axes[1].set_title(f'Setelah Pencerahan (b={nilai_b})')

plt.show()
```
